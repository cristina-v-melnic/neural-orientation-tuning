# Demo neural-orientation-selectivity 

## 1. Main functionality 

### 1.1. Create connected neuron objects

 - Specify the connectivity type: "weight" or "number" - based
 - Specify the desired PO of the postsynaptic neuron (rad)
 - Class ConnectedNeuron(connectivity_type = "x", desired_PO = "np.pi/4")

```python
# Import classes for circuit connectivity
# and for spike-based LIF integration. 

from ConnectedNeuron import *
from SpikyLIF import *
from TunedResponse import *
```

```python
Weight_neuron = ConnectedNeuron(connectivity_type="weight")
```

```python
Number_neuron = ConnectedNeuron(connectivity_type="number", to_plot=True)
```

### 1.3. Visualise the background firing activity of the postsynaptic neuron

```python
Weight_background_LIF = SpikyLIF(Weight_neuron)
```

```python
Number_background_LIF = SpikyLIF(Number_neuron, to_plot=True)
```

### 1.2. Present a bar oriented at angle $\theta = \frac{\pi}{4}$ rad

- See the response of one of the afferents, using ConnectedNeuron and TunedResponse classes.
- See the response of the postsynaptic neuron using ConnectedNeuron, SpikyLIF and TunedResponse classes.

```python
Response_1_bar = TunedResponse(Number_neuron, trials=1, nr_bars=1, theta = [np.pi/4])
```

```python
Response_1_bar.plot_postsynaptic_curve()
```

```python
# The response firing rate of the postsynaptic neuron.
print(Response_1_bar.fs_out)
```

### 1.3. Present a range of stimuli over multiple trials $\theta = [0, \pi]$ (rad) 

- Set nr of trials
- Get the postsynaptic tuning curve

```python
Response_range_bar = TunedResponse(Number_neuron, trials=5, nr_bars=10)
```

```python
Response_range_bar.plot_postsynaptic_curve()
```

```python
# Firing rates for all trials and stimuli.
print(Response_range_bar.fs_out)
```
